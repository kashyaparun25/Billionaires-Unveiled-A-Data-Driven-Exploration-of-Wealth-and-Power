## Countries with the highest concentration of billionaires.


```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(18,6))
sns.countplot( data=df, x="country", order=df['country'].value_counts().index)
plt.xticks(rotation=90)
plt.ylabel("No of Billionaires", fontsize=15)
plt.xlabel("Countries", fontsize=15)
plt.title("No of Billionaires vs. Countries", fontsize=20)
plt.show()
```

## The ages of individuals who are billionaires.

```python
plt.figure(figsize = (10,6))
sns.histplot(data = df, x = 'age', color = 'brown')
plt.title('Age distribution of the billionaires',fontsize=20)
plt.show()
```

## Who holds the title of the youngest and oldest individuals in terms of age?


```python
df.loc[df.age == df.age.min()]
```

```python
df.loc[df.age == df.age.max()]
```

```python
df['age_group'] = pd.cut(df['age'],bins = [18, 30, 50, 70, 100], include_lowest = True,labels = ['18-30', '30-50', '50-70','70+'])
df.head()
```

```python
df['age_group'].value_counts().plot(kind='bar', figsize=(7,5))
plt.title('Billionaires Count vs. Age Group',fontsize=16)
plt.xlabel("Age Group", fontsize=15)
plt.ylabel("Billionaires Count", fontsize=15)
plt.show()
```

## Which industry boasts the largest number of billionaires?

```python
df['networth']= df['networth'].str.replace('$', '').str.replace(' B', '').astype(float)
```

```python
plt.figure(figsize=(15,8))
sns.barplot(data=df, x = "networth", y = "industry", order = df['industry'].value_counts().index, estimator = sum , errwidth = 0)
plt.xticks(rotation=90)
plt.xlabel("Networth", fontsize=15)
plt.ylabel("Industries", fontsize=15)
plt.show()
```

## Examine the potential correlation between wealth and age.

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['age'], df['networth'])
plt.xlabel('Age')
plt.ylabel('Net Worth (in $Billion)')
plt.title('Age vs. Net Worth')
plt.grid(True)
```

## The leading billionaires originating from India.


```python
indians = df[df['country'] == 'India']
```

```python
top_5_indians = indians.sort_values('networth',ascending=False).head()
```

```python
print(top_5_indians)
```

```python
plt.figure(figsize=(10, 6))

# Plot the top 5 Indians' net worth
plt.bar(top_5_indians['name'], top_5_indians['networth'])
plt.xlabel('Billionaire Name')
plt.ylabel('Net Worth (in billions)')
plt.title('Top 5 Indians by Net Worth')

# Display the industry as labels on top of each bar
for i, v in enumerate(top_5_indians['industry']):
    plt.text(i, top_5_indians['networth'].iloc[i] + 1, str(v), ha='center', va='bottom')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## The leading billionaires originating from the United States.

```python
americans = df[df['country'] == 'United States']
```

```python
top_5_americans = americans.sort_values('networth',ascending=False).head()
plt.figure(figsize=(10, 6))

# Plot the top 5 Americans' net worth
plt.bar(top_5_americans['name'], top_5_americans['networth'])
plt.xlabel('Billionaire Name')
plt.ylabel('Net Worth (in billions)')
plt.title('Top 5 Americans by Net Worth')

# Display the industry as labels on top of each bar
for i, v in enumerate(top_5_americans['industry']):
    plt.text(i, top_5_americans['networth'].iloc[i] + 1, str(v), ha='center', va='bottom')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
